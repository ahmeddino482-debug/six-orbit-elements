```python
import numpy as np

def gibbs_method(r1,r2,r3,mu = 398600):
    
    r1 = np.array(r1)
    r2 = np.array(r2)
    r3 = np.array(r3)

    mag_r1 = np.linalg.norm(r1)
    mag_r2 = np.linalg.norm(r2)
    mag_r3 = np.linalg.norm(r3)

    c12 = np.cross(r1,r2)
    c23 = np.cross(r2,r3)
    c31 = np.cross(r3,r1)

    N = (mag_r1 * c23) + (mag_r2 * c31) + (mag_r3 * c12)
    D = c12 + c23 + c31
    S = r1*(mag_r2 - mag_r3) + r2*(mag_r3 - mag_r1) + r3*(mag_r1 - mag_r2)

    N_mag = np.linalg.norm(N)
    D_mag = np.linalg.norm(D)

    v1 = (np.sqrt(mu / (N_mag * D_mag))) * (((np.cross(D , r1)) / mag_r1) + S)

    L = np.sqrt(mu / (N_mag * D_mag))  
    M = ((np.cross(D , r2)) / mag_r2) + S
    v2 = L * M
    
    v3 = (np.sqrt(mu / (N_mag * D_mag))) * (((np.cross(D , r3)) / mag_r3) + S)

    v2_mag = np.linalg.norm(v2)

    h1 = np.cross(r2,v2)
    h_mag = np.linalg.norm(h1)

    E_1 = (v2_mag * v2_mag) / 2
    E_2 = mu / mag_r2

    E = E_1 - E_2

    a = -mu / (2*E)

    e_1 = np.cross(v2 , h1) / mu
    e_2 = r2 / mag_r2

    e_vec = e_1 - e_2

    e_mag = np.linalg.norm(e_vec)

    I = np.array([1,0,0])
    J = np.array([0,1,0])
    K = np.array([0,0,1])

    i = np.arccos(np.dot(h1 , K) / h_mag)
    i = np.degrees(i)

    N_Ax = np.cross(K , h1)

    if N_Ax[1] > 0:
        R_A = np.arccos(np.dot(N_Ax , I) / np.linalg.norm(N_Ax))
        R_A = np.degrees(R_A)
    else:
        R_A = 360 - (np.arccos(np.dot(N_Ax , I) / np.linalg.norm(N_Ax)))
        R_A = np.degrees(R_A)
    
    if e_vec[2] > 0:
        W = np.arccos(np.dot(N_Ax , e_vec) / (np.linalg.norm(N_Ax) * e_mag))
        W = np.degrees(W)
    else:
        W = 360 - np.arccos(np.dot(N_Ax , e_vec) / (np.linalg.norm(N_Ax) * e_mag))
        W = np.degrees(W)

    if e_vec[2] > 0:
        T_A = np.arccos(np.dot(e_vec , r2) / (mag_r2 * e_mag))
        T_A = np.degrees(T_A)
    else:
        T_A = 360 - np.arccos(np.dot(e_vec , r2) / (mag_r2 * e_mag))
        T_A = np.degrees(T_A)
    

    return a , e_mag, i,h_mag , R_A, W, T_A 
```
